```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import random
```

```python
ticker_symbol = "VCB"
start_date = "01/01/2023"  # Specify the start date
end_date = "31/12/2023"  # Specify the end date
url = f"https://s.cafef.vn/tin-doanh-nghiep/{ticker_symbol}/Event.chn"
base_url = "https://s.cafef.vn"
date_format = "%d/%m/%Y"
start_date = datetime.strptime(start_date, date_format).date()
end_date = datetime.strptime(end_date, date_format).date()
result = []
```

```python
# Send a GET request to the URL
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

# Find divEvents containing all li
div_events = soup.find("div", id="divEvents")
# Find all the li elements
news_items = div_events.find_all("li")
# Remove all li having wrong time frame
# print(news_items);
```

```python
class Article:
    def __init__(self, title, content, date):
        self.title = title
        self.content = content
        self.date = date
        self.point = round(random.random(),1) # Default level
    def __str__(self):
        return f"Title: {self.title}\nDate: {self.date}\nContent: {self.content}\nPoint: {self.point}\n"
```

```python
for new_item in news_items:
    new_link = new_item.find("a")["href"]
    news_title = new_item.find("a").text.strip()
    # dd/mm/yyyy hh/ss
    created_date_time = new_item.find("span").text.strip()
    # dd/mm/yyyy
    created_date = created_date_time.split(" ")[0]
    created_date = datetime.strptime(created_date, date_format).date()
    
    # Check if upload time is valid or not
    if (created_date < start_date or created_date > end_date):
        # print(created_date, start_date, end_date)
        continue
    
    # Send new request
    new_response = requests.get(base_url+new_link)
    new_soup = BeautifulSoup(new_response.content, "html.parser")
    
    # Find newscontent div to extract content
    news_content_div = new_soup.find("div", id="newscontent") 
    news_content = news_content_div.get_text(strip=True)
    
    if (len(news_content) < 300):
        continue
    temp_article = Article(news_title, news_content, created_date)
    result.append(temp_article)
    
```

```python
for article in result:
    print(article)
```
